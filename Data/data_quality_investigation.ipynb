```python
df_test = df.groupby(['Reference', 'Assay', 'REGION_NOTES', 'Sequence'], dropna=False).agg({'pIC50': list, 'Sequence ID': list}).reset_index()
df_test.head()

#group by ['Reference', 'Assay', 'REGION_NOTES', 'Sequence'] and agg pIC50 and Sequence ID into lists

print(len(df_test[df_test['pIC50'].apply(len) > 1]))
#only display rows where pIC50 list has more than 1 value
df_test = df_test[df_test['pIC50'].apply(len) > 1]
df_test
```

```python
#calculate the average absolute difference between the pIC50 values
avg_diff = []
for pIC50_list in df_test['pIC50']:
    if (len(pIC50_list) == 2):
        #print(np.mean(np.abs(np.diff(pIC50_list))))
        avg_diff.append(np.mean(np.abs(np.diff(pIC50_list))))

mean_avg_diff = np.mean(avg_diff)
print(avg_diff)
print(mean_avg_diff)
```
